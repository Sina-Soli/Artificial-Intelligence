# C

```python
# 20% Test & 80% Training
Training_Data_Number = int(np.round_((80/100)*num_rows))
Testimg_Data_Number = int(num_rows - Training_Data_Number)
print(Training_Data_Number)
print("***********************")
print(Testimg_Data_Number)
```

```python
#Shuffling And Split generated Data to Train and Test
trained =file_np[0:Training_Data_Number,:]
#np.random.shuffle(trained)
Tested = file_np[Training_Data_Number:,:]
#np.random.shuffle(Tested) 
```

```python
X_train = trained[:, 0:num_cols-1]
Y_train = trained[:, num_cols-1]
Y_train = np.reshape(Y_train,(Training_Data_Number,1))
print("X_train is: \n" , Y_train)
#print("***********************")
#print(X_train[0])
#print(Y_train[0])
print("***********************")
X_test = Tested[:, 0:num_cols-1]
Y_test = Tested[:, num_cols-1]
Y_test = np.reshape(Y_test,(Testimg_Data_Number,1))
print("X_test is: \n" , X_test)
#print("***********************")
#print(X_test[0])
#print(Y_test[0])
```

# D

```python
mean =  np.mean(X_train) 
standard_deviation = np.std( X_train)
X_train_scaled = (X_train -mean) /standard_deviation
```

```python
#print(X_train_scaled)
```

```python
mean =  np.mean(X_test) 
standard_deviation = np.std( X_test)
X_test_scaled = (X_test -mean) /standard_deviation
#print(X_test_scaled)
```

# E

```python
# Creating our model
from keras import models
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import layers
from keras.metrics import mean_absolute_error

def mlp():
    model = models.Sequential()

    model.add(layers.Dense(256, activation='relu', input_shape=(13,)))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(16, activation='relu'))
    
    model.add(layers.Dense(1))
    
    #model.summary()
    model.compile(optimizer='rmsprop', loss="mse", metrics=['mean_absolute_error'])
    return model
```

# F

```python
from sklearn.model_selection import KFold
folds = KFold(n_splits=5, random_state=None)
mae = []
 
for i, j in folds.split(X_train_scaled):
    model = mlp()

    x_fold_train, x_fold_valid = X_train_scaled[i,:], X_train_scaled[j,:]
    y_fold_train, y_fold_valid = Y_train[i], Y_train[j]

    model.fit(x_fold_train, y_fold_train, epochs=4, batch_size=1)
    predict = model.predict( x_fold_valid).reshape( y_fold_valid.shape)

    mae.append(mean_absolute_error( y_fold_valid, predict))
```

```python
test_mse, test_mae = model.evaluate(X_test_scaled, Y_test)
print('Loss of testset (MSE):', test_mse)
print('Metric of testset (MAE):', test_mae)
```
